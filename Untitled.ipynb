```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN

# 读取数据
df = pd.read_csv(r"D:\composite_0602\initial_dataset\jia1-s3-1 nodes.csv")

# 使用概率大于0.5的候选点进行聚类
threshold = 0.5
candidates = df[df['is_singularity'] > threshold].copy()

if len(candidates) > 0:
    # 基于x,y坐标进行DBSCAN聚类，eps=0.05可根据数据分布调整
    coords = candidates[['x', 'y']].values
    clustering = DBSCAN(eps=0.05, min_samples=1).fit(coords)
    candidates['cluster'] = clustering.labels_

    # 在每个聚类中找到概率最大的点
    max_prob_idx = candidates.groupby('cluster')['is_singularity'].idxmax()
    final_singularity_nodes = set(candidates.loc[max_prob_idx, 'node_id'])

    # 生成新列
    df['is_singularity_final'] = df['node_id'].isin(final_singularity_nodes).astype(int)
else:
    df['is_singularity_final'] = 0

# 保存结果
# df.to_csv('jia1-s3-1_nodes_with_final_singularity.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN

# 读取数据（请根据实际文件路径修改）
df = pd.read_csv(r"D:\composite_0602\initial_dataset\jia1-s3-2 nodes.csv")

# ---------- 确定最终奇异点 ----------
threshold = 0.5
candidates = df[df['is_singularity'] > threshold].copy()

if len(candidates) > 0:
    coords = candidates[['x', 'y']].values
    # eps 可根据坐标范围调整，这里 0.05 适合 [0,1] 区间
    clustering = DBSCAN(eps=0.05, min_samples=1).fit(coords)
    candidates['cluster'] = clustering.labels_
    max_prob_idx = candidates.groupby('cluster')['is_singularity'].idxmax()
    final_nodes = set(candidates.loc[max_prob_idx, 'node_id'])
    df['is_singularity_final'] = df['node_id'].isin(final_nodes).astype(int)
else:
    df['is_singularity_final'] = 0

# ---------- 可视化 ----------
plt.figure(figsize=(10, 8))

# 绘制所有节点，颜色表示原始奇异点概率
sc = plt.scatter(df['x'], df['y'], c=df['is_singularity'], 
                 cmap='viridis', s=15, alpha=0.7, edgecolors='none')
plt.colorbar(sc, label='Singularity Probability')

# 标记最终确定的奇异点
final_points = df[df['is_singularity_final'] == 1]
plt.scatter(final_points['x'], final_points['y'], 
            marker='*', s=50, c='red', edgecolors='black', 
            linewidth=1, label='Final Singularity', zorder=5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Singularity Probability Field and Determined Singularities')
plt.legend()
plt.grid(alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.savefig('singularity_visualization.png', dpi=150)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN

def assign_final_psl(df, prob_col='is_on_psl', coord_cols=('x', 'y'),
                     node_id_col='node_id', threshold=0.5, eps=0.05, min_samples=1):
    """
    根据 is_on_psl 概率值和坐标聚类，在每个聚类中选择概率最大的节点作为最终 PSL 点。

    Parameters
    ----------
    df : pd.DataFrame
    prob_col : str, default='is_on_psl'
    coord_cols : tuple, default=('x','y')
    node_id_col : str, default='node_id'
    threshold : float, default=0.5
    eps : float, default=0.05
    min_samples : int, default=1

    Returns
    -------
    pd.DataFrame
        新增一列 `is_on_psl_final`。
    """
    df_out = df.copy()
    candidates = df_out[df_out[prob_col] > threshold].copy()
    if len(candidates) == 0:
        df_out['is_on_psl_final'] = 0
        return df_out

    coords = candidates[list(coord_cols)].values
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(coords)
    candidates['cluster'] = clustering.labels_

    max_prob_idx = candidates.groupby('cluster')[prob_col].idxmax()
    final_nodes = set(candidates.loc[max_prob_idx, node_id_col])

    df_out['is_on_psl_final'] = df_out[node_id_col].isin(final_nodes).astype(int)
    return df_out
```

```python
df = pd.read_csv(r"D:\composite_0602\initial_dataset\jia1-s3-2 nodes.csv")
df = assign_final_psl(df, prob_col='is_on_psl', threshold=0.5, eps=0.05)
print(df[['node_id', 'is_on_psl', 'is_on_psl_final']].head())
```

```python
import os
import re
import pandas as pd
import numpy as np
from glob import glob

# 配置路径
data_dir = r"D:\composite_0602\initial_dataset"  # 存放所有csv文件的目录
output_dir = "./merged_data"  # 合并后的输出目录
os.makedirs(output_dir, exist_ok=True)

# 记录错误信息
errors = []

def parse_file_pairs(files):
    """根据文件名后缀(-1/-2 或 .1/.2)将文件配对"""
    pairs = {}
    pattern = re.compile(r'(.*?)(?:-1|\.1)(?=\.csv$)')
    for f in files:
        base = re.sub(r'(-1|-2|\.1|\.2)(?=\.csv$)', '', f)
        # 提取序号 1 或 2
        suffix_match = re.search(r'(-1|-2|\.1|\.2)(?=\.csv$)', f)
        if suffix_match:
            suffix = suffix_match.group(0)
            if base not in pairs:
                pairs[base] = {}
            if '-1' in suffix or '.1' in suffix:
                pairs[base]['1'] = f
            else:
                pairs[base]['2'] = f
    # 只保留完整配对的
    return {base: pair for base, pair in pairs.items() if '1' in pair and '2' in pair}

def merge_dataframes(df1, df2, id_col, skip_cols=None):
    """合并两个DataFrame，处理占位符'/'，检查一致性"""
    if skip_cols is None:
        skip_cols = []
    # 确保id列存在
    if id_col not in df1.columns or id_col not in df2.columns:
        raise ValueError(f"Missing id column {id_col}")
    
    # 按id合并，保留所有行（outer）
    merged = pd.merge(df1, df2, on=id_col, how='outer', suffixes=('_1', '_2'))
    
    # 收集所有需要合并的列（除去id和skip_cols）
    all_cols = set(df1.columns) | set(df2.columns)
    common_cols = all_cols - {id_col} - set(skip_cols)
    
    result = pd.DataFrame()
    result[id_col] = merged[id_col]
    
    for col in common_cols:
        col1 = col + '_1'
        col2 = col + '_2'
        # 如果该列只在一边存在
        if col1 not in merged.columns:
            result[col] = merged[col2]
            continue
        if col2 not in merged.columns:
            result[col] = merged[col1]
            continue
        
        # 获取两边的值
        vals1 = merged[col1].fillna('/').replace('', '/')
        vals2 = merged[col2].fillna('/').replace('', '/')
        
        # 合并逻辑
        def merge_cell(v1, v2):
            if v1 == '/' and v2 == '/':
                return '/'
            if v1 == '/':
                return v2
            if v2 == '/':
                return v1
            if str(v1) != str(v2):
                # 记录冲突
                errors.append(f"Conflict in column '{col}': {v1} vs {v2}")
                return v1  # 默认取第一个，可调整
            return v1
        
        result[col] = [merge_cell(v1, v2) for v1, v2 in zip(vals1, vals2)]
    
    # 处理skip_cols：直接取df1中的值（假设两边一致，否则报错）
    for col in skip_cols:
        if col in df1.columns and col in df2.columns:
            # 比较是否一致
            temp = pd.merge(df1[[id_col, col]], df2[[id_col, col]], on=id_col, how='outer', suffixes=('_1', '_2'))
            for idx, row in temp.iterrows():
                v1 = row.get(col+'_1', '/')
                v2 = row.get(col+'_2', '/')
                if str(v1) != str(v2) and v1 != '/' and v2 != '/':
                    errors.append(f"Conflict in skip column '{col}': {v1} vs {v2}")
            result[col] = df1.set_index(id_col)[col].reindex(result[id_col]).fillna(df2.set_index(id_col)[col]).fillna('/')
        elif col in df1.columns:
            result[col] = df1.set_index(id_col)[col].reindex(result[id_col]).fillna('/')
        elif col in df2.columns:
            result[col] = df2.set_index(id_col)[col].reindex(result[id_col]).fillna('/')
        else:
            result[col] = '/'
    
    return result

def process_files(group_name, file1, file2, file_type):
    """处理一组文件（nodes或edges）"""
    print(f"Processing {group_name} {file_type}...")
    df1 = pd.read_csv(file1, keep_default_na=False, na_values=[''])
    df2 = pd.read_csv(file2, keep_default_na=False, na_values=[''])
    
    # 检查必要列
    if file_type == 'nodes':
        id_col = 'node_id'
        skip_cols = ['is_singularity', 'is_on_psl']
    else:  # edges
        id_col = 'edge_id'
        skip_cols = []  # edges中可能没有特殊列，根据实际调整
    
    merged_df = merge_dataframes(df1, df2, id_col, skip_cols)
    
    # 输出合并文件
    out_filename = f"{group_name}_merged_{file_type}.csv"
    out_path = os.path.join(output_dir, out_filename)
    merged_df.to_csv(out_path, index=False)
    print(f"Saved to {out_path}")

def main():
    # 获取所有csv文件
    all_files = glob(os.path.join(data_dir, "*.csv"))
    # 分离nodes和edges
    nodes_files = [f for f in all_files if "nodes" in os.path.basename(f)]
    edges_files = [f for f in all_files if "edges" in os.path.basename(f)]
    
    # 配对nodes
    nodes_pairs = parse_file_pairs([os.path.basename(f) for f in nodes_files])
    for base, pair in nodes_pairs.items():
        file1 = os.path.join(data_dir, pair['1'])
        file2 = os.path.join(data_dir, pair['2'])
        process_files(base, file1, file2, 'nodes')
    
    # 配对edges
    edges_pairs = parse_file_pairs([os.path.basename(f) for f in edges_files])
    for base, pair in edges_pairs.items():
        file1 = os.path.join(data_dir, pair['1'])
        file2 = os.path.join(data_dir, pair['2'])
        process_files(base, file1, file2, 'edges')
    
    # 打印错误记录
    if errors:
        print("\n=== Errors encountered ===")
        for err in set(errors):  # 去重
            print(err)
    else:
        print("\nNo errors found. All data consistent.")

if __name__ == "__main__":
    main()
```
